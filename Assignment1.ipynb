**Problem 1 (based on 1.1.2)**
Use the Intermediate Value Theorem to find an interval of length one that contains a solution of the equation 

The first step to solving an equation is to verify that a root exists. One way to ensure
this is to bracket the root: to find an interval [a,b] on the real line for which one of the pair
{f (a),f (b)} is positive and the other is negative

**Problem 3 (based on Comp 1.2.2)**
Apply Fixed-Point iteration to find the solution of each equation to eight correct decimal places. Your solution should include a table of intermediate approximations xi along with the errors ei = |x - r| and the ratios wi/ei-1, similar to the table on page 37. 

Remark: In order to find ei one first needs to calculate r.

```python
import math

#a) x^5 + x = 1
# Define the function f(x) and the corresponding g(x)
def f(x):
    return x**5 + x - 1

def g(x):
    return 1 - (1 * x**5)

# Define the tolerance for eight decimal places
tolerance = 1e-9

# Initialize variables
x_i = 0.25  # Initial guess
r = None  # To be determined

# Use FPI to find the fixed point r
while True:
    x_i_plus_1 = g(x_i)
    if abs(x_i_plus_1 - x_i) < tolerance:
        r = x_i_plus_1
        break
    x_i = x_i_plus_1

# Initialize variables for the table
n = 0.00
tabledic = {
    'i': [],
    'xi': [],
    'g(xi)': [],
    'ei=|xi - r|': [],
    'ei / ei-1': [],  # Changed the column name
}

# Calculate errors and ratios using FPI
while True:
    n += 1
    x_i_plus_1 = g(x_i)
    error = abs(x_i_plus_1 - r)
    ratio = error / tabledic['ei=|xi - r|'][-1] if 'ei=|xi - r|' in tabledic and len(tabledic['ei=|xi - r|']) > 1 and tabledic['ei=|xi - r|'][-1] != 0 else 0  # Calculate the ratio (wi/ei-1)

    x_i = x_i_plus_1

    # Add values to the table dictionary
    tabledic['i'].append(n)
    tabledic['xi'].append(x_i)
    tabledic['g(xi)'].append(g(x_i))
    tabledic['ei=|xi - r|'].append(error)
    tabledic['ei / ei-1'].append(ratio)  # Changed the column name

    if error < tolerance:
        break

# Print the table
for key, values in tabledic.items():
    print(f"{key}: {values}")

# Print the final approximation with eight decimal places
print(f"The solution to x^5 + x = 1 with eight decimal places of accuracy is: {x_i:.8f}")
```

**Problem 4 (based on 1.2.8)**
Use Theorem 1.6 to determine whether Fixed-Point Iteration of g(x) is locally convergent to the given fixed point r in the following three cases:

(a) g(x) = (2x-1)/(x^2), r = 1,
(b) g(x) = cos(x) + π + 1, r = π,
(c) g(x) = (e^2x) - 1, r = 0.

In the cases where the Fixed-Point Iteration converges, determine also the convergence rate.

```python
import sympy as sp

# Define the symbol and the fixed point
x = sp.Symbol('x')
r_values = [1, sp.pi, 0]

# Define the g(x) functions for each case
g_functions = [
    (2*x - 1) / (x**2),
    sp.cos(x) + sp.pi + 1,
    sp.exp(2*x) - 1
]

for i, g in enumerate(g_functions):
    r = r_values[i]

    # Calculate the derivative of g(x)
    g_prime = sp.diff(g, x)

    # Evaluate g'(r)
    convergence_condition = abs(g_prime.subs(x, r))

    print(f"Case {i + 1}:")
    print(f"g(x) = {g}")
    print(f"Fixed point r = {r}")
    print(f"g'(x) = {g_prime}")
    print(f"|g'({r})| = {convergence_condition}")

    if convergence_condition < 1:
        p = convergence_condition
        print(f"Convergent: Fixed-Point Iteration is locally convergent.")
        print(f"Convergence rate p = |g'({r})| = {p:.6f}\n")
    else:
        print("Not convergent: Fixed-Point Iteration may not converge.")
        print(f"Convergence condition |g'({r})| >= 1 is not satisfied.\n")
```

**Problem 6 (Comp 1.3.4)**
(a) Use the sensitivity formula for roots Δr ≈ (εg(r))/(f'(r) 
    (formula (1.21) in the textbook) to approximate the root near 3 
    of fε(x) = (1 + ε)x^3 - 3x^2 + x - 3 for a constant ε. 
(b) Setting ε = 10^-3, find the actual root and compare with part (a). 

```python
# Define the function f(x)
def f(x, epsilon):
    return (1 + epsilon) * x**3 - 3*x**2 + x - 3

# Define the derivative of f(x)
def f_derivative(x, epsilon):
    return 3*(1 + epsilon)*x**2 - 6*x + 1

# Set the value of epsilon
epsilon_value = 1e-3

# Set the root you want to approximate near
root_guess = 3.0

# Maximum number of iterations (to prevent infinite looping)
max_iterations = 100

# Tolerance for convergence
tolerance = 1e-8

# Initialize the approximation
approximate_root = root_guess

for i in range(max_iterations):
    # Calculate the sensitivity
    delta_r = (epsilon_value * f(approximate_root, epsilon_value)) / f_derivative(approximate_root, epsilon_value)
    
    # Update the approximate root
    new_approximation = approximate_root - delta_r
    
    # Calculate the absolute error
    absolute_error = abs(new_approximation - approximate_root)
    
    # Check for convergence
    if absolute_error < tolerance:
        break
    
    # Update the approximation
    approximate_root = new_approximation

print(f"Approximate root with ε = {epsilon_value}: {approximate_root:.6f}")

# Calculate the actual root using numerical methods (e.g., Newton-Raphson)
def newton_raphson(f, f_prime, x0, epsilon, max_iterations):
    x = x0
    for _ in range(max_iterations):
        f_value = f(x, epsilon)
        f_prime_value = f_prime(x, epsilon)
        if abs(f_value) < epsilon:
            return x
        x = x - f_value / f_prime_value
    return None

actual_root = newton_raphson(f, f_derivative, root_guess, tolerance, max_iterations)
if actual_root is not None:
    print(f"Actual root with ε = {epsilon_value}: {actual_root:.6f}")
else:
    print("Newton-Raphson did not converge to the actual root.")
```
